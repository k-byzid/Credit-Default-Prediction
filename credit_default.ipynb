```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os
import math
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, precision_score, recall_score
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
```

# Auxillary Functions

```python
def path_loader(path, view = False):
    '''
        This function takes the path to dataset and returns
        the path for training and test data, and feature
        descriptions.
        
        view = False : do not view the paths
        view = True : view the paths
        
    '''
    for directory_path, _ , directory_files in os.walk(path):
        path_all = []
        for files in directory_files:
            path_all.append(os.path.join(directory_path, files))
        for i, path in enumerate(path_all):
            if 'test' in path:
                path_testing = path
            elif 'training' in path:
                path_training = path
            elif 'Dictionary' in path:
                path_help = path
    
    if view == True:
        print('Training Data Path: {}\nTest Data Path: {}\nData Dictionary Path: {}'.format(path_training, path_testing, path_help))
        
    return path_training, path_testing, path_help
```

```python
def col_name_changer(dataframe, name_dictionary):
    '''
        This function takes a name_dictionary and a
        dataframe, it changes the names of the columns
        according to the key-values given in the dictionary.
        
    '''
    dataframe = dataframe.rename(columns = name_dictionary)
    
    return dataframe
```

```python
def count_visualizer(dataset, col_name):
    '''
        This is used to visualize the count of unique
        values in a given pandas column
        
    '''
    data = dataset
    fig, ax = plt.subplots()
    ax.hist(data[col_name])
    ax.set_title('Frequency of Unique Values')
    ax.set_xlabel('Unique Value Label')
    ax.set_ylabel('Count')
```

```python
def model_evaluation(y_pred, y_test):
    '''
        This function is used to evaluate the models based on
        y_test and the predicted y_pred.
    
    '''
    cm = confusion_matrix(y_pred, y_test)
    
    print('MAE: ', metrics.mean_absolute_error(y_test, y_pred))
    print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
    print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print("Precision Score: ",precision_score(y_test, y_pred, average = 'weighted'))
    print("Recall Score: ",recall_score(y_test, y_pred, average = 'weighted'))
    print("f1_score: ",f1_score(y_test, y_pred, average = 'weighted'))
    print('\n\033[1m' +'Confusion Matrix : '+ '\033[0m')
    sns.heatmap(cm, cmap = 'OrRd', annot = True, fmt='d')
```

```python
def detect_outliers(dataset, features, sensitivity_factor):
    '''
        This function takes a given dataset and features for
        outlier detection. The average deviation is first
        calculated which is then multiplied by the factor
        to give a suitable range for outlier detecting.
        
        sensitivity_factor > 3 would result in less number of
        outliers.
        
        sensitivity_factor suggested range = [> 1 and < 3]
        
    '''
    outlier_index = []
    outlier_list = list(np.zeros(len(features), dtype = int))
    
    for i, col in enumerate(features):
        Q1 = dataset.describe()[col][4]
        Q2 = dataset.describe()[col][5]
        Q3 = dataset.describe()[col][6]
        
        avg_deviation = ((Q3 - Q2) + (Q2 - Q1))/2
        
        outlier_range = avg_deviation * sensitivity_factor
        outlier_list[i] = dataset[(dataset[col] > Q2 + outlier_range) | (dataset[col] < Q2 - outlier_range)].index
    
    outlier_index.extend(np.intersect1d(outlier_list[0], outlier_list[1]))
    print('Number of data: {}\nNumber of outliers found: {}\nNumber of remaining data: {}'.format(len(dataset), 
                                                                                                  len(outlier_index), 
                                                                                                  len(dataset) - len(outlier_index)))
        
    return outlier_index
```

```python
def model_pipeline(dataset, test_size, knn_args, dt_args, view_data_shape = False):
    '''
        This function takes in a dataset and divides it according to a 
        given test_size. These are fit into the ML models.
        
        set view_data_shape = True to see the dataset shape after division.
        
        Models : Decision Tree and K-Nearest-Neighbour Classifier.
        
        knn_args:
            "algorithm", "leaf_size", "metric", "n_jobs", "n_neighbors", "p", "weights"
            
        dt_args:
            "criterion", "min_samples_leaf", "min_samples_split", "random_state", "splitter"
    
    '''
    
    x_train, x_test, y_train, y_test = train_test_split(dataset[features], dataset[target], test_size = test_size, 
                                                        random_state = 39, shuffle = True)
    
    if view_data_shape == True:
        print("Shape of x_train: {}\nShape of y_train: {}\nShape of x_test: {}\nShape of y_test: {}\n".format(x_train.shape,
                                                                                                              y_train.shape,
                                                                                                              x_test.shape,
                                                                                                              y_test.shape))
        
    # Call the KNN model and evaluate the accuracy on test data
    KNN_model = KNeighborsClassifier(**knn_args)

    KNN_model.fit(x_train, y_train)
    y_pred_KNN = KNN_model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred_KNN)
    print(f"Accuracy for KNN Model is: {accuracy * 100}")
        
    # Call the Decision Tree model and evaluate the accuracy on test data
    DT_model = tree.DecisionTreeClassifier(**dt_args)

    DT_model.fit(x_train, y_train)
    y_pred_DT = DT_model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred_DT)
    print(f"Accuracy for Decision Tree is: {accuracy * 100}")
        
    return y_pred_KNN, y_pred_DT, y_test    
```

# Dataset Procurement and Preprocessing

### Procurement

```python
path = "Dataset"

path_training, path_testing, path_help = path_loader(path, view = True)

dataset_description = pd.read_csv(path_help, header = 1)
train = pd.read_csv(path_training)
test = pd.read_csv(path_testing)
```

```python
'''
    The columns chosen for dropping were chosen based on
    specific application based testing and information requirement
    for decent accuracy

''' 

columns_to_drop = [0, 5, 6, 9, 11]
train = train.drop(train.columns[columns_to_drop], axis = 1)
test = test.drop(test.columns[columns_to_drop], axis = 1)
```

```python
col_dictionary = {'SeriousDlqin2yrs': 'target',
                  'RevolvingUtilizationOfUnsecuredLines': 'balance-limit-ratio',
                  'NumberOfTime30-59DaysPastDueNotWorse': 'late30-59',
                  'NumberOfOpenCreditLinesAndLoans': 'purchase_count',
                  'NumberOfTimes90DaysLate': 'late>90',
                  'NumberOfTime60-89DaysPastDueNotWorse': 'late60-89'}

train = col_name_changer(train, col_dictionary)
test = col_name_changer(test, col_dictionary)

columns = train.columns
features = columns[1:]
target = columns[0]
```

### Data Analysis and Processing

```python
train.describe()
```

```python
#Check for missing/zero data:

missing_count_train = (train[features].isnull().sum())/(len(train[features]))
missing_count_test = (test[features].isnull().sum())/(len(test[features]))

print(f"Number of features: {len(features)}")
print(f"Number of features without missing/zero data in train dataset: {missing_count_train.isnull().count()}")
print(f"Number of features without missing/zero data in test dataset: {missing_count_test.isnull().count()}")
```

```python
# Visualize the distribution of the unique target values

count_visualizer(train, target)
```

```python
# There are too few target = 1 cases, so we can duplicate a few rows

multiplier_factor = 3

for i in range(multiplier_factor):
    indexes = train[(train[target] == 1)].index
    indexes = random.sample(list(indexes), len(train[(train[target] == 1)].index))
    indexes = pd.Index(indexes, dtype = 'Int64')
    rows_to_duplicate = train.loc[indexes]
    train = pd.concat([train, rows_to_duplicate], ignore_index=True)
```

```python
# Drop rows with outlier or artifact values

outlier_features = [features[0], features[3]]
outliers_to_drop = detect_outliers(train, outlier_features, sensitivity_factor = 2.5)

train = train.drop(outliers_to_drop, axis = 0).reset_index(drop = True)
```
